```python
import os, fnmatch

def dirfiles(img_path,img_ext):
    img_names = fnmatch.filter(sorted(os.listdir(img_path)),img_ext)
    return img_names
```

```python
import numpy as np
from tqdm.auto import tqdm
#crea la particion
p_train = 0.70  # 70% del total de los datos son training
p_val   = 0.20  # 20% del total de los datos son de validación
carpeta="/content/drive/MyDrive/TesisDataCarrosEtiquetas/"
d = dirfiles(carpeta,'*.jpg')
m = len(d) # número de imagenes

nt = round(p_train*m)
#nv = round(p_val*nt)
nv = round(p_val*m)
print('m: ', m)
print('nt: ', nt)
print('nv: ', nv)

ii = np.vstack([np.zeros((nt,1)),np.ones((nv,1)),2*np.ones((m-nt-nv,1))])
target = ['train','val','test']

for i in tqdm(range(m)):
  fname            = d[i]
  tname            = fname[:-3]+'txt'
  ti               = target[int(ii[i])]
  source_file      = carpeta+fname
  destination_file = 'TesisDataCarros/images/'+ti 
  os.system('cp ' + source_file + ' ' + destination_file) 
  source_file      = carpeta+tname
  destination_file = 'TesisDataCarros/labels/'+ti 
  os.system('cp ' + source_file + ' ' + destination_file) 

print('Training: ', nt-nv)
print('     Val: ', nv)
print(' Testing: ',m-nt)
```
